```python
import numpy as np
import pandas as pd

numerical = ['Pclass', 'Age', 'Family', 'Fare', 'Fare/Family+1']
categorical = ['Sex', 'Embarked', 'Cabin_is_null', 'Honorific', 'Deck', 'TicketPrefix', 'TicketNumber']#,  'PclassFemale']#, 'PclassMale', 'PclassSex']

def preprocess(path):
    df = pd.read_csv(path)
    #df['Age_is_null'] = df['Age'].isnull()
    df['Cabin_is_null'] = df['Cabin'].isnull()
    df['Age'] = df['Age'].fillna(df['Name'].map(noAge))
    df['Honorific'] = df['Name'].str.split(', ').str[1].str.split('. ').str[0]
    df['Family'] = df['SibSp'] + df['Parch']
    df['Fare/Family+1'] = df['Fare'] / (df['Family']+1)
    df['Deck'] = df['Cabin'].str[0]
    df['TicketPrefix'] = df['Ticket'].str.extract(r'([A-Za-z]+)')
    df['TicketPrefix'] = df['TicketPrefix'].str.upper()
    #df['TicketPrefix'] = df['TicketPrefix'].isnull()
    df['TicketNumber'] = df['Ticket'].str.extract(r'([0-9]+)')
    df['TicketNumber'] = df['TicketNumber'].str.len()
    
    df['PclassFemale'] = (df['Pclass']!=3) & (df['Sex']=='female')
    #df = pd.concat([df, dummies], axis=1)
    #print(df.isnull().sum())
    df.fillna(df.median(numeric_only=True), inplace=True)
    df[categorical] = df[categorical].fillna('unknown')
    df[categorical] = df[categorical].astype('category')

    df[['Fare', 'Family', 'Fare/Family+1']] = np.log1p(df[['Fare', 'Family', 'Fare/Family+1']])
    return df

df_train = preprocess("../data/train.csv")
df_test = preprocess("../data/test.csv")
```

```python
import pandas as pd
import numpy as np
import optuna
optuna.logging.set_verbosity(optuna.logging.WARNING)
from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from tqdm.notebook import tqdm

X = df_train[numerical + categorical]
y = df_train['Perished']

cat_features = [X.columns.get_loc(col) for col in categorical]  # CatBoost uses column indices for categorical features

N_TRIAL = 100
pbar = tqdm(total=N_TRIAL)

def objective(trial):
    params = {
        'loss_function': 'Logloss',
        'eval_metric': 'Logloss',
        'depth': trial.suggest_int('depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1.0, 10.0),
        'iterations': trial.suggest_int('iterations', 50, 300),
        'random_seed': 42,
        'verbose': 0,
        'early_stopping_rounds': 30,
        #'task_type': 'GPU',  # Uncomment if GPU is available
    }

    accuracies = []
    y_valids = []
    y_preds = []

    sfk = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    for train_idx, valid_idx in sfk.split(X, y):
        X_train, X_valid = X.iloc[train_idx], X.iloc[valid_idx]
        y_train, y_valid = y.iloc[train_idx], y.iloc[valid_idx]

        train_pool = Pool(X_train, y_train, cat_features=cat_features)
        valid_pool = Pool(X_valid, y_valid, cat_features=cat_features)

        model = CatBoostClassifier(**params)
        model.fit(train_pool, eval_set=valid_pool)

        y_pred = model.predict(X_valid)
        acc = accuracy_score(y_valid, y_pred)
        accuracies.append(acc)
        y_valids.extend(y_valid)
        y_preds.extend(y_pred)

    if not hasattr(objective, "best_score") or np.mean(accuracies) > objective.best_score:
        objective.best_score = np.mean(accuracies)
        objective.best_y_valid = y_valids
        objective.best_y_pred = y_preds

    pbar.update(1)
    return 1 - np.mean(accuracies)

# Optunaによるハイパーパラメータ探索
study = optuna.create_study(direction='minimize', sampler=optuna.samplers.TPESampler(seed=42))
study.optimize(objective, n_trials=N_TRIAL)
pbar.close()
```

```python
# 最適なハイパーパラメータの表示
print("Best trial:")
print(study.best_trial.params)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from catboost import CatBoostClassifier, Pool
from sklearn.metrics import confusion_matrix

# CatBoost用のPoolを作成（categoricalは列インデックスで指定）
cat_features_indices = [X.columns.get_loc(col) for col in categorical]
pool = Pool(X, label=y, cat_features=cat_features_indices)

# best_trialのパラメータでCatBoostモデルを学習
params = study.best_trial.params.copy()
params.update({
    "loss_function": "Logloss",
    "verbose": 0,
    "random_seed": 42,
})
model = CatBoostClassifier(**params)
model.fit(pool)

# 精度表示
print("Accuracy:", accuracy_score(objective.best_y_valid, objective.best_y_pred))

# 混同行列
tn, fp, fn, tp = confusion_matrix(objective.best_y_valid, objective.best_y_pred).ravel()
df_conf = pd.DataFrame(
    [[tp, fp],
     [fn, tn]],
    index=["Perished (True)", "Survived (True)"],
    columns=["Perished (Pred)", "Survived (Pred)"]
)

# 可視化
f, ax = plt.subplots(1, 2, figsize=(16, 6), facecolor='gray')
sns.heatmap(df_conf, annot=True, fmt='d', cmap='Blues', ax=ax[0])
ax[0].set_title("Confusion Matrix (TP/FP/FN/TN)")

# 特徴量重要度をプロット
importances = model.get_feature_importance(pool, type='FeatureImportance', )
sns.barplot(x=importances, y=X.columns, ax=ax[1])
ax[1].set_title("Feature Importance (CatBoost)")

plt.tight_layout()
plt.show()
```
